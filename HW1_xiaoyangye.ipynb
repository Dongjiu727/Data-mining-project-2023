```python
# Import necessay libraries or packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load and check the Data
df1 = pd.read_csv("/content/Cereals.csv")
display(df1.head(5))
display(df1.info())
```

```python
# Horizontal bar chart to visulize the calories of different cereals in df1 database

# Get the values
name_of_cereals = df1["name"]
calories_of_cereals = df1["calories"]

#Create a bar chart
plt.figure(figsize=(30, 40))
plt.barh(name_of_cereals,calories_of_cereals,color= "b")

#
plt.xlabel('The calories of cereals',fontsize = 15)
plt.ylabel('Name of cereals',fontsize = 15)
plt.title('The calories of different cereals',fontsize = 17)
plt.grid(linestyle='--')
plt.show()
```

```python
# Line chart to show the changes of work hours over years
#Load and check the Data
df2 = pd.read_csv("/content/CanadianWorkHours.csv")
display(df2.head(5))
display(df2.info())

#Create the line chart
Years = df2["Year"]
working_hours = df2["Hours"]

plt.figure(figsize=(10, 10))
plt.plot(Years,working_hours,color='tab:blue')
plt.xlabel('Years')
plt.ylabel('Working hours')
plt.title('The working hours over years in a special area')
plt.show()
```

```python
#Pie chart to show the Manufacturer of cereal (mfr)of Cereals in df1

# Get the manufacturer counts
mfr_counts = df1['mfr'].value_counts()

# Extract labels and values
labels = mfr_counts.index
values = mfr_counts.values

# Create a pie chart
plt.figure(figsize=(8, 6))
plt.pie(values, labels=labels, autopct='%1.1f%%', textprops=dict(color="k"),labeldistance = 1.08,startangle=30)

plt.title("Manufacturer Distribution of Cereals")
plt.legend(labels, loc="upper left")

plt.axis('equal')
plt.show()
```

```python
#Scatter plots to show the distribution of calories vs carbo of different cereals

#Extract the data
carbo = df1["carbo"]

#Create a scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(calories_of_cereals, carbo, alpha=0.8)
plt.xlabel('calories of cereals' , fontsize = 12)
plt.ylabel('Carbo of cereals' , fontsize = 12)
plt.title("Calories vs Carbo in different cereals")
plt.grid(linestyle='-')
plt.show()
```
